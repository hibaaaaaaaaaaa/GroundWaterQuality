```python
#question1

import pandas as pd
import math as mt
from functools import reduce
import numpy as np

df=pd.read_excel('WaterQuality.xlsx') #Importation du fichier excel
dfp1=pd.read_excel('Paramètres1.xlsx') #dataframe contenant 'Si'et 'wi' des paramètres
dfp1.index=['Si','wi']
dfp1
```

                              Moyenne arithmetique pondérée (WAWQI)

```python
#question2

def Wi():    #fonction permettant le calcul du Wi 
    list1=[]
    for i in dfp1.iloc[1,:]:
        S=0
        for j in dfp1.iloc[1,0:]:
            S=S+j
        Wi=i/S
        list1.append(Wi)
    dfp1.loc["Wi",:]=list1
Wi()   
dfp1
```

```python
#question2

Wi()   # l'appel de la fonction WI()

def WAWQI():   # Calcul du WAWQI() et l'insertion des résultats dans df
    list1=[]
    for i in range(0,18):
        list2=[]
        for key, value in df.iteritems(): 
            if key=="pH":
              v=value[i]
              Qi=(v-8.5)/(6.5-8.5)
            elif key=="FC" or key=="Year" or key=="POINTS": continue
            else:
              v=value[i]
              Qi=(v/dfp1.loc["Si",key])*100
            SII=Qi*dfp1.loc["Wi",key]
            list2.append(SII)
        list1.append(sum(list2))
    df.loc[:,"WAWQI"]=list1   
    
WAWQI()
df
```

```python
#question3
def interpretation(): #interpretation des résultats selon la valeur du WAWQI
    list1=[]
    for i in df.loc[:,"WAWQI"]:
       if i<50:
           q="Excellent Water"
       elif i>=50 and i<=100:
           q="Good Water"
       elif i>=101 and i<=200:
           q="Poor Water"
       elif i>=201 and i<=300:
           q="Very Poor water"
       else:
           q="Unsuitable for drinking"
       list1.append(q)
    df.loc[:,"Quality"]=list1
    
interpretation()
df
```

                            Moyenne géométrique pondérée (WGWQI)

```python
#question 5

Wi() #appel de la fonction WI()

def WGWQI(): #Définition d'une fonction permettant le calcul du WGWQI et l'insertion des résultats dans df1
    list1=[]
    for i in range(0,18):
        list2=[]
        for key, value in df1.iteritems(): 
            if key=="pH":
              v=value[i]
              Qi=(v-8.5)/(6.5-8.5)
            elif key=="FC" or key=="Year" or key=="POINTS": continue
            else:
              v=value[i]
              Qi=(v/dfp1.loc["Si",key])*100
            qiwi=mt.pow(Qi,dfp1.loc["Wi",key])
            list2.append(qiwi)
        list1.append(reduce(lambda x,y:x*y,list2))
    df1.loc[:,"WGWQI"]=list1  
WGWQI()
df1
```

```python
#question6
def interpretation(): #interprétation des valeurs du WGWQI
    list1=[]
    for i in df1.loc[:,"WGWQI"]:
       if i>=90 and i<=100:
           q="Excellent"
       elif i>=70 and i<=89:
           q="Good"
       elif i>=50 and i<=69:
           q="Medium"
       elif i>=25 and i<=49:
           q="Bad"
       else:
           q="Very Bad"
       list1.append(q)
    df1.loc[:,"Quality"]=list1
interpretation()
df1
```

                       L'indice d'Oregon de Qualité de l'eau (OWQI)

```python
#question8

dfp1=pd.read_excel('Paramètres1.xlsx') #importation du fichier contenant wi et si des paramètres
dfp1.index=['Si','wi']

list1=[594,5]
dfp1.loc[:,"FC"]=list1 # ajout d'une colonne pour 'FC'

list2=["DCO","CONDUCTIVITE","MES","TURBIDITE","NITRITE"]
for i in list2:
    dfp1.pop(i) #suppression des paramètres de la liste 2 
    
    
Wi() #appel de la fonction WI() 
dfp1
```

```python
#question8
def func(a,b): #définiton d'une fonction qui permet de retourner l'infini en cas de division par 0
    try:
        return a/b
    except ZeroDivisionError as e:
        return np.inf
```

```python
#question8 
def OWQI(): #calcul du OWQI
    list1=[]
    for i in range(0,18):
        list2=[]
        for key, value in df2.iteritems(): 
            if key=="pH":
              v=value[i]
              Qi=(v-8.5)/(6.5-8.5)
            elif key=="DCO" or key=="Year" or key=="POINTS" or key=="CONDUCTIVITE" or key=="MES" or key=="TURBIDITE" or key=="NITRITE": continue
            else:
              v=value[i]
              Qi=(v/dfp1.loc["Si",key])*100
            SII=Qi*dfp1.loc["Wi",key]
            list2.append(func(1,mt.pow(SII,2)))
        list1.append(mt.sqrt(8/sum(list2)))
    df2.loc[:,"OWQI"]=list1
    
OWQI()
df2
```

```python
#question9
def interpretation(): #interprétation des résultats
    list1=[]
    for i in df2.loc[:,"OWQI"]:
       if i>=90 and i<=100:
           q="Excellent"
       elif i>=85 and i<=89:
           q="Good"
       elif i>=80 and i<=84:
           q="Fair"
       elif i>=60 and i<=79:
           q="Poor"
       else:
           q="Very Poor"
       list1.append(q)
    df2.loc[:,"Quality"]=list1
    
interpretation()
df2
```
